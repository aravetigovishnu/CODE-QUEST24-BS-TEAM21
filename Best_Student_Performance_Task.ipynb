```python
import numpy as np
import pandas as pd
from sklearn.model_selection import train_test_split
from sklearn.ensemble import RandomForestRegressor
from sklearn.metrics import accuracy_score
```

```python
student = pd.read_csv('student_performance.csv')
```

```python
X = student.drop(['Student_name', 'Overall_Score'], axis = 1)
y = student['Overall_Score']
```

```python
X_train, X_test, y_train, y_test = train_test_split(X, y, test_size = 0.3, random_state = 42)
```

```python
model = RandomForestRegressor(n_estimators = 100, max_depth = 5, random_state = 42)
model.fit(X_train, y_train)
```

```python
y_pred = model.predict(X_test)
```

```python
from sklearn.metrics import mean_squared_error, r2_score

mse = mean_squared_error(y_test, y_pred)
r2 = r2_score(y_test, y_pred)

print("Mean Squared Error:", mse)
print("R-squared Score:", r2)
```

```python
training_accuracy = model.score(X_train, y_train)
testing_accuracy = model.score(X_test, y_test)

print("Training Accuracy:", training_accuracy)
print("Testing Accuracy:", testing_accuracy)
```

```python
x_new = model.predict([[1001,2024,7.5,3,4,5,6,1,6.7,86.5,5]])
print(x_new)
```

```python
predicted_scores = model.predict(X)


student['Predicted_Score'] = predicted_scores


student_ranked = student.sort_values(by='Predicted_Score', ascending=False)


top_students = student_ranked.groupby('Year').head(3)

top_students['Rank'] = top_students.groupby('Year')['Predicted_Score'].rank(method='first', ascending=False)

years = top_students['Year'].unique()

for year in sorted(years):
    print(f"\n{year}:")
    year_students = top_students[top_students['Year'] == year]
    for _, row in year_students.iterrows():
        print(f"{row['Student_name']: <20} {row['Predicted_Score']: <10} {int(row['Rank'])}")
```
